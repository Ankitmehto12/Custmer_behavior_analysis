```python
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly':7,
    'Monthly':30,
    'Quarterly':90,
    'Bi-Weekly':14,
    'Annually':365,
    'Every 3 Months': 90    
}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```
